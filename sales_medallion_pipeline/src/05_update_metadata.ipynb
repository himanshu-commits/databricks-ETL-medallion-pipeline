```python
df_incremental = spark.table(
    "staging.sales_incremental"
)

max_ts = (
    df_incremental
    .selectExpr(
        "max(updated_at) as max_ts"
    )
    .collect()[0]["max_ts"]
)

if max_ts is None:

    print(
        "No new data found. Watermark not updated."
    )

else:

    spark.sql(f"""
    
    UPDATE metadata.pipeline_metadata
    
    SET watermark =
    '{max_ts}'
    
    WHERE pipeline_name =
    'sales_pipeline'
    
    """)

    print(
        f"Watermark updated to {max_ts}"
    )
```
