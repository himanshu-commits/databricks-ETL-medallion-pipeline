```python
%sql

MERGE INTO silver.sales s

USING staging.sales_incremental b

ON s.customer_id = b.customer_id

WHEN MATCHED THEN
UPDATE SET
s.amount = b.amount,
s.region = b.region,
s.updated_at = b.updated_at

WHEN NOT MATCHED THEN
INSERT (
customer_id,
amount,
region,
updated_at
)
VALUES (
b.customer_id,
b.amount,
b.region,
b.updated_at
)
```
